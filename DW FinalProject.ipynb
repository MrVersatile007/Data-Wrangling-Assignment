```python
########## 1. IMPORTING DATASET  FROM URL

import pandas as pd
import requests

url = "https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv"

data = pd.read_csv(url)
```

```python
########## 2. HIGH LEVEL DATA UNDERSTANDING
```

```python
#A. FIND NUMBER OF ROWS AND COLUMNS IN DATASET
print('Number of Rows : ',data.shape[0])
print('Number of Cols : ',data.shape[1])
```

```python
#B. DATA TYPES OF COLUMNS
data.dtypes
```

```python
#C. INFO AND DESCRIBE OF DATA IN DATA FRAME
data.info()
```

```python
data.describe(include = 'all')
```

```python
########## 3. LOW LEVEL DATA UNDERSTANDING
```

```python
#A. FIND COUNT OF UNIQUE VALUES IN LOCATION COLUMN
print('the total count of unique values are : ',data.location.value_counts().shape[0])
```

```python
#DETAIL IDEA ABOUT THE COUNT OF THOSE UNIQUE VALUES
data.location.value_counts()
```

```python
#B. Find which continent has maximum frequency using values counts
data.continent.value_counts().head(1)
```

```python
#C. Find maximum & mean value in 'total_cases'.
print('maximum in "total_cases" : ', data.total_cases.max())
print('mean in "total_cases"    : ', data.total_cases.mean())
```

```python
#D. Find 25%, 50% & 75% quartile value in 'total_deaths'.
data.describe().total_deaths.loc[["25%", "50%", "75%"]]
```

```python
#E.  Find which continent has maximum 'human_development_index'.
HDI_filter = data.human_development_index.max()
set(data.loc[data.human_development_index == HDI_filter, 'continent'])
```

```python
#F. Find which continent has minimum 'gdp_per_capita'
GDP_filter = data.gdp_per_capita.min()
set(data.loc[data.gdp_per_capita == GDP_filter, 'continent'])
```

```python
########## 4. FILTER DATAFRAME WITH ONLY THIS COLUMNS
data = data.loc[:,['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
data
```

```python
########## 5. DATA CLEANING
```

```python
#A. REMOVE ALL DUPLICATE OBSERVATIONS
data.drop_duplicates()
```

```python
#B. FIND MISSING VALUES IN ALL COLUMNS
data.isna().sum()
```

```python
#C. Remove all observations where continent column value is missing
data.dropna(subset=['continent'], inplace= True)
data.reset_index(drop = True, inplace = True)
```

```python
#D. Fill all missing values with 0

data.fillna(0)
```

```python
########## 6. DATE TIME FORMAT
data.date
```

```python
#A. Convert date column in datetime format using pandas.to_datetime
data["date"]= pd.to_datetime(data["date"])
data.date
```

```python
#B. Create new column month after extracting month data from date column.
data['month'] = pd.DatetimeIndex(data['date']).month
data.month
```

```python
########## 7. DATA AGGREGATION
```

```python
#A. Find max value in all columns using groupby function on 'continent' column
data.groupby('continent').max()
data.reset_index(inplace = True)
```

```python
#B. Store the result in a new dataframe named 'df_groupby'.
df_groupby = data.groupby('continent').max()
df_groupby['continent'] = df_groupby.index
df_groupby
```

```python
########## 8. FEATURE ENGINEERING 
```

```python
#A. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby.total_deaths/ df_groupby.total_cases
df_groupby
```

```python
########## 9. Data Visualization
import seaborn as sns
```

```python
#A. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
sns.distplot(x = df_groupby['gdp_per_capita'] , kde= False)
```

```python
#B. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
sns.scatterplot(data= df_groupby, x= 'total_cases', y= 'gdp_per_capita', hue= 'continent')
```

```python
#C. Plot Pairplot on df_groupby dataset.
sns.pairplot(df_groupby, hue = 'continent')
```

```python
#D. Plot a bar plot of 'continent' column with 'total_cases' .
sns.catplot(data = df_groupby, x= 'continent' , y= 'total_cases', kind= 'bar')
```
